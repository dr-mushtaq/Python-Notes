## **8.5 Bitwise Operators**<a name='bit_operators'></a>

````python
x = 5  # Binary representation: 0101
y = 3  # Binary representation: 0011

print(x & y)   # Bitwise AND:    0001 (Decimal: 1)
print(x | y)   # Bitwise OR:     0111 (Decimal: 7)
print(x ^ y)   # Bitwise XOR:    0110 (Decimal: 6)
print(~x)      # Bitwise NOT:    -6 (Decimal: -6)
print(x << 1)  # Left Shift:     1010 (Decimal: 10)
print(x >> 1)  # Right Shift:    0010 (Decimal: 2)
````

## **8.6 Identity operators**<a name='identify_operators'></a>

````python
# Integers (small integers are often stored as singletons)
a = 5
b = 5
print(a is b)  # True (they might share the same memory location)

# Strings (immutable, often reused if the same value exists)
c = "hello"
d = "hello"
print(c is d)  # True (they might share the same memory location)

# Lists (mutable, each list has its own memory location)
e = [1, 2, 3]
f = [1, 2, 3]
print(e is f)  # False (even though they have the same values, they are different objects)

# Checking with `id()`
print(id(a))  # Same for `a` and `b` if they share memory
print(id(c))  # Same for `c` and `d` if they share memory
print(id(e))  # Different for each `list` object
````

````python
a = [1,2,3]
b = [1,2,3]
print(a is b)     # False (different objects)
print(a == b)     # True (same values)

x = None
print(x is None)  # True
````

A loop is used to repeat a block of code multiple times until a condition is met.

There are two types of loops in Python.

 **For Loop** and **While Loop**.

## **11.1 `For` Loops**<a name='for_loop'></a>
For loops iterate over a given sequence. Here is an example:

````python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
 print(fruit)
````

````python
primes = [2, 3, 5, 7]
for prime in primes:
    print(prime)
````

For loops can iterate over a sequence of numbers using the **`range`** and **`xrange`** functions. The difference between range and xrange is that the range function returns a **`new list`** with numbers of that specified range, whereas xrange returns **`an iterator`**, which is more efficient. (Python 3 uses the range function, which acts like xrange). Note that the range function is zero based.

````python
# Prints out the numbers 0,1,2,3,4
for x in range(5):
    print(x)
````

````python
# Prints out 3,4,5
for x in range(3, 6):
    print(x)
````

````python
# Prints out 3,5,7
for x in range(3, 8, 2):
    print(x)
````

## **11.2 `while loops`**<a name='while_loop'></a>
While loops repeat as long as a certain boolean condition is met. For example:



````python
# Prints out 0,1,2,3,4
count = 0
while count < 5:
    print(count)
    count += 1  # This is the same as count = count + 1
````

````python
# While loop
count = 1
while count<5:
    print(count)
    count += 1
````

````python
# While loop with else block
count = 1
while count<5:
    print(count)
    count += 1
else:
    print('End of else')
````

````python
# Infinite While Loop
count = 1
while True:
    print(count, end=' ')
    count += 1
````

## **11.3 `Break` and `Continue` statements**<a name='break_continue_statement'></a>
break is used to exit a for loop or a while loop, whereas continue is used to skip the current block, and return to the **`for`** or **`while`** statement. A few examples:

````python
# Prints out 0,1,2,3,4
count = 0
while True:
    print(count)
    count += 1
    if count >= 5:
        break
````

````python
# Prints out only odd numbers - 1,3,5,7,9
for x in range(10):
    # Check if x is even
    if x % 2 == 0:
        continue
    print(x)
````

If break statement is executed inside for loop then the **`else`** part is skipped. Note that **`else`** part is executed even if there is a continue statement.

````python
# Prints out 0,1,2,3,4 and then it prints "count value reached 5"

count=0
while(count<5):
    print(count)
    count +=1
else:
    print("count value reached %d" %(count))

# Prints out 1,2,3,4
for i in range(1, 10):
    if(i%5==0):
        break
    print(i)
else:
    print("this is not printed because for loop is terminated because of break but not due to fail in condition")
````

# **Section 17- File Handling**<a name='file_handling'></a>

##**17.1 Access Modes for Reading a File**

The key function for working with files in Python is the **`open()`** function.
The **`open()`** function takes two parameters; `filename`, and `mode`. **There are four different methods (modes) for opening a file**:

| Mode | Name | Description |
| :--- | :--- | :--- |
| `"r"` | **Read** | **Default.** Opens for reading; errors if the file is missing. |
| `"a"` | **Append** | Adds to the end of the file; creates it if it doesn't exist. |
| `"w"` | **Write** | Overwrites the file; creates it if it doesn't exist. |
| `"x"` | **Create** | Creates a new file; **errors** if the file already exists. |


**Syntax:**

````python
f = open("demofile.txt")
````

## **17.2 Opening and Closing Files**<a name='open_close_file'></a>

The first step in file handling is opening a file for reading, writing, or both. Python's `open()` function is used to open a file, specifying the file name and mode (`'r'` for reading, `'w'` for writing, `'a'` for appending, and more). It's essential to close the file using the `close()` method once you're done with it to free up system resources.

````python
# Open a file for reading
file = open('example.txt', 'r')
````

## **17.3 Reading from Files**<a name='read_files'></a>
